```python
import pandas as pd
import os

import shutil
```

```python
#Create the data for the positive (Github)
FILE_PATH="covid-chestxray-dataset-master/covid-chestxray-dataset-master/metadata.csv"
IMAGES_PATH="covid-chestxray-dataset-master/covid-chestxray-dataset-master/images"
```

```python
df=pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

```python
TARGET_DIR="Dataset/Covid"
if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("Covid Folder Created")
```

```python
cnt=0
for(i,row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        filename=row["filename"]
        image_path=os.path.join(IMAGES_PATH,filename)
        image_copy_path=os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)
        #print("Moving images",cnt)
        cnt+=1
#print(cnt)
```

```python
#Sampling of Images From Kaggle
import random
KAGGLE_FILE_PATH="archive/chest_xray/test/NORMAL"
TARGET_NORMAL_DIR="Dataset/Normal"
```

```python
image_names=os.listdir(KAGGLE_FILE_PATH)
```

```python
random.shuffle(image_names)
```

```python
for i in range(196):
    image_name=image_names[i]
    image_path=os.path.join(KAGGLE_FILE_PATH,image_name)
    target_path=os.path.join(TARGET_NORMAL_DIR,image_name)
    shutil.copy2(image_path,target_path)
    print("Copying images",i)
```
